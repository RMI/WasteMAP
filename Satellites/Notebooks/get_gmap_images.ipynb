```python
import pandas as pd
import leafmap
from samgeo import tms_to_geotiff
from samgeo.text_sam import LangSAM
import rasterio

locs = pd.read_csv('C:/Users/hughr/OneDrive/Documents/RMI/METER/cnlandfillapr11.csv')

locs_1 = locs.iloc[:100, :]
locs_2 = locs.iloc[100:200, :]
locs_3 = locs.iloc[200:300, :]
locs_4 = locs.iloc[300:400, :]
locs_5 = locs.iloc[400:500, :]
locs_6 = locs.iloc[500:, :]

print(len(locs))
print(len(locs_1) + len(locs_2) + len(locs_3) + len(locs_4) + len(locs_5) + len(locs_6))

locs_list = [locs_1, locs_2, locs_3, locs_4, locs_5, locs_6]
```

```python
for l in locs_list:
    for idx, row in l.iterrows():
        # Create a 3 km x 3 km Polygon around each point 
        lat, lon = row['lat'], row['lon']
        bbox = [lon - .01, lat - .01, lon + .01, lat + .01]
        image = 'C:/Users/hughr/OneDrive/Documents/RMI/METER/gmap_imgs/' + str(idx) + '.tif'
        #image='test.tif'
        tms_to_geotiff(output=image, bbox=bbox, resolution=1, source="Satellite", overwrite=True)
```
